```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_node("fact_treatment")

G.add_node("dim_patient")
G.add_node("dim_doctor")
G.add_node("dim_hospital")
G.add_node("dim_time")

G.add_node("dim_city")
G.add_node("dim_department")
G.add_node("dim_specialization")
G.add_node("dim_month")
G.add_node("dim_year")

G.add_edge("fact_treatment", "dim_patient")
G.add_edge("fact_treatment", "dim_doctor")
G.add_edge("fact_treatment", "dim_hospital")
G.add_edge("fact_treatment", "dim_time")

G.add_edge("dim_patient", "dim_city")
G.add_edge("dim_doctor", "dim_specialization")
G.add_edge("dim_hospital", "dim_department")
G.add_edge("dim_time", "dim_month")
G.add_edge("dim_month", "dim_year")

pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(10, 7))
nx.draw(G, pos, with_labels=True, node_size=3000, font_size=9)
plt.title("Hospital Snowflake Schema")
plt.show()
```
